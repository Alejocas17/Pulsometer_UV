```python
import serial
from time import time
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import json
```

```python
t_inicial = time()
intervalo=15;
intervalo2=0;

serial_port = "COM3";
baud_rate = 9600;
write_to_file_path = "muestra.txt";
output_file = open(write_to_file_path, "w+");

ser = serial.Serial(serial_port, baud_rate);
flag=True;
i=0;
```

```python
while flag:
    line = ser.readline();
    line = line.decode("utf-8");
    output_file.write(line);       
    i=i+1;
    if(time()-t_inicial>intervalo):
        intervalo2=time()-t_inicial;
        flag=False;
    
    
```
